```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset("titanic")
```

```python
df.head()
```

```python
plt.figure(figsize=(8,6))

sns.boxplot(x='sex',
         y='age',
         hue='survived',
         data=df)
plt.title('Boxplot of age with sex and survived')

plt.show()
         
```

```python
sns.violinplot(x='sex',
               y='age',
               hue='survived',
               data=df)

plt.title('violinplot of age with sex and survival')

plt.show()
```
